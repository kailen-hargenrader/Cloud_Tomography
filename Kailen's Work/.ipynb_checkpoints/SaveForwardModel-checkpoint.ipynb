## Simulating Radiances

This example covers the simualation of synthetic measurements using pyshdom. This is a simple and streamlined example.

```python
#imports

import at3d
import numpy as np
import xarray as xr
from collections import OrderedDict
import pylab as py
```

### Make the RTE grid and medium microphysics.

This time we are going to include both cloud and rayleigh scatter. 
The fundamental thing that must be defined first is the `rte_grid`, which sets the domain size. We have an easy choice here from the medium but in the inverse problem this must be selected as the initial step.

```python
cloud_scatterer = at3d.util.load_from_csv('../data/synthetic_cloud_fields/jpl_les/rico32x37x26.txt',
                                           density='lwc',origin=(0.0,0.0))

#load atmosphere
atmosphere = xr.open_dataset('../data/ancillary/AFGL_summer_mid_lat.nc')
#subset the atmosphere, choose only the bottom four km.
reduced_atmosphere = atmosphere.sel({'z': atmosphere.coords['z'].data[atmosphere.coords['z'].data <= 4.0]})
#merge the atmosphere and cloud z coordinates
merged_z_coordinate = at3d.grid.combine_z_coordinates([reduced_atmosphere,cloud_scatterer])

# define the property grid - which is equivalent to the base RTE grid
rte_grid = at3d.grid.make_grid(cloud_scatterer.x.diff('x')[0],cloud_scatterer.x.data.size,
                          cloud_scatterer.y.diff('y')[0],cloud_scatterer.y.data.size,
                          np.append(0,cloud_scatterer.z.data))
```

```python
print(rte_grid)
```

```python
print(cloud_scatterer)
```

```python
#finish defining microphysics because we can.

cloud_scatterer_on_rte_grid = at3d.grid.resample_onto_grid(rte_grid, cloud_scatterer)

#We choose a gamma size distribution and therefore need to define a 'veff' variable.
size_distribution_function = at3d.size_distribution.gamma

cloud_scatterer_on_rte_grid['veff'] = (cloud_scatterer_on_rte_grid.reff.dims,
                                       np.full_like(cloud_scatterer_on_rte_grid.reff.data, fill_value=0.1))
```

```python
print(cloud_scatterer_on_rte_grid)
```

### Define Sensors

This is based on AirMSPI - like measurements. 9 observations in an arc.

Two perspective cameras either ground-based or out of an aircraft window.

and eMAS observations - nadir multi-spectral measurements.

By default, the resolution arguments for the sensors are actually spacing between rays. The radiance calculation itself is still just a single idealized ray. To more accurately model a wide pixel field of view we can model several sub-pixel-rays in each pixel.

```python
sensor_dict = at3d.containers.SensorsDict()
```

```python
sensor_zenith_list = [75.0,67.5,60.0,52,45.6,35,26.1,10.0]*2 + [0.0]
sensor_azimuth_list = [90]*8 + [-90]*8 +[0.0]
for zenith,azimuth in zip(sensor_zenith_list,sensor_azimuth_list):
    sensor_dict.add_sensor('MSPI', 
            at3d.sensor.orthographic_projection(0.86, cloud_scatterer,0.02,0.02, azimuth, zenith,
                                     altitude='TOA', stokes='I')
                          )
```

```python
# add some perspective cameras
sensor_dict.add_sensor('SideViewCamera', 
        at3d.sensor.perspective_projection(0.86, 20.0,100,100,[5.0, 0.32, 1.0],[0.01,0.32,1.0],[0,0,1],
                                 stokes='I')
                      )

sensor_dict.add_sensor('GroundCamera', 
        at3d.sensor.perspective_projection(0.86, 80.0,120,120,[0.32, 0.32, 0.001],[0.32,0.32,1.0],[0,1,0],
                                 stokes='I')
                      )
```

```python
for wavelength in [1.65, 2.2, 3.7]:
    sensor_dict.add_sensor('eMAS', 
            at3d.sensor.orthographic_projection(wavelength, cloud_scatterer,0.02,0.02, azimuth, zenith,
                                     altitude='TOA', stokes='I',
                                                  )
                          )
```

Here we make a sensor that now integrates over the radiance field in each pixel's field of view using gauss-legendre quadrature.

```python
for wavelength in [1.65]:
    sensor_dict.add_sensor('eMAS_SubPixel', 
            at3d.sensor.orthographic_projection(wavelength, cloud_scatterer,0.02,0.02, azimuth, zenith,
                                     altitude='TOA', stokes='I',
                                    sub_pixel_ray_args={
                                        'method':at3d.sensor.gaussian,
                                        'degree':3
                                    })
                          )
```

```python
wavelengths = sensor_dict.get_unique_solvers()
```

```python
print(wavelengths)
```

### get optical property generators

Note that we assume a monochromatic model here.

```python
mie_mono_tables = OrderedDict()
for wavelength in wavelengths:
    mie_mono_tables[wavelength] = at3d.mie.get_mono_table(
        'Water',(wavelength,wavelength),
        max_integration_radius=65.0,
        minimum_effective_radius=0.1,
        relative_dir='../mie_tables',
        verbose=False
    )
```

```python
optical_property_generator = at3d.medium.OpticalPropertyGenerator(
    'cloud', 
    mie_mono_tables,
    size_distribution_function,
    reff=np.linspace(5.0,30.0,30),
    veff=np.linspace(0.03,0.2,9),
)
optical_properties = optical_property_generator(cloud_scatterer_on_rte_grid)
```

```python
print(optical_properties.keys())
```

```python
# one function to generate rayleigh scattering.
rayleigh_scattering = at3d.rayleigh.to_grid(wavelengths,atmosphere,rte_grid)
```

## Define Solvers

Define solvers last based on the sensor's spectral information.

```python
# NB IF YOU REDEFINE THE SENSORS BUT KEEP THE SAME SET OF SOLVERS 
# THERE IS NO NEED TO REDEFINE THE SOLVERS YOU CAN SIMPLY RERUN
# THE CELL BELOW WITHOUT NEEDING TO RERUN THE RTE SOLUTION.

solvers_dict = at3d.containers.SolversDict()
# note we could set solver dependent surfaces / sources / numerical_config here
# just as we have got solver dependent optical properties.

for wavelength in sensor_dict.get_unique_solvers():
    medium = {
        'cloud': optical_properties[wavelength],
        'rayleigh':rayleigh_scattering[wavelength]
     }
    config = at3d.configuration.get_config()
    solvers_dict.add_solver(
        wavelength,
        at3d.solver.RTE(
            numerical_params=config,
            surface=at3d.surface.lambertian(0.0),
            source=at3d.source.solar(wavelength, 0.5,0.0),
            medium=medium,
            num_stokes=1#sensor_dict.get_minimum_stokes()[wavelength],
        )                   
   )
```

```python
print(len(sensor_dict["MSPI"]["sensor_list"]))
print(len(solvers_dict))
```

```python
%%time
# solve the 4 RTEs in parallel AND get the measurements.
sensor_dict.get_measurements(solvers_dict, n_jobs=4, verbose=True)
```

```python
###Compare Same Cloud with Different REFF###
def get_sensors_from_data(data):
    
    #finish defining microphysics because we can.
    
    boosted_cloud_scatterer_on_rte_grid = at3d.grid.resample_onto_grid(rte_grid, data)
    boosted_cloud_scatterer_on_rte_grid['veff'] = (boosted_cloud_scatterer_on_rte_grid.reff.dims,
                                           np.full_like(boosted_cloud_scatterer_on_rte_grid.reff.data, fill_value=0.1))
    
    #Get Sensors
    boosted_sensor_dict = at3d.containers.SensorsDict()
    for zenith,azimuth in zip(sensor_zenith_list,sensor_azimuth_list):
        boosted_sensor_dict.add_sensor('MSPI', 
                at3d.sensor.orthographic_projection(0.86, data,0.02,0.02, azimuth, zenith,
                                         altitude='TOA', stokes='I')
                              )
    
    boosted_sensor_dict.add_sensor('SideViewCamera', 
            at3d.sensor.perspective_projection(0.86, 20.0,100,100,[5.0, 0.32, 1.0],[0.01,0.32,1.0],[0,0,1],
                                     stokes='I')
                          )
    
    boosted_sensor_dict.add_sensor('GroundCamera', 
            at3d.sensor.perspective_projection(0.86, 80.0,120,120,[0.32, 0.32, 0.001],[0.32,0.32,1.0],[0,1,0],
                                     stokes='I')
                          )
    
    for wavelength in [1.65, 2.2, 3.7]:
        boosted_sensor_dict.add_sensor('eMAS', 
                at3d.sensor.orthographic_projection(wavelength, data,0.02,0.02, azimuth, zenith,
                                         altitude='TOA', stokes='I',
                                                      )
                              )
    
    for wavelength in [1.65]:
        boosted_sensor_dict.add_sensor('eMAS_SubPixel', 
                at3d.sensor.orthographic_projection(wavelength, data,0.02,0.02, azimuth, zenith,
                                         altitude='TOA', stokes='I',
                                        sub_pixel_ray_args={
                                            'method':at3d.sensor.gaussian,
                                            'degree':3
                                        })
                              )
    #Get optical properties
    boosted_optical_properties = optical_property_generator(boosted_cloud_scatterer_on_rte_grid)
    
    # NB IF YOU REDEFINE THE SENSORS BUT KEEP THE SAME SET OF SOLVERS 
    # THERE IS NO NEED TO REDEFINE THE SOLVERS YOU CAN SIMPLY RERUN
    # THE CELL BELOW WITHOUT NEEDING TO RERUN THE RTE SOLUTION.
    
    boosted_solvers_dict = at3d.containers.SolversDict()
    # note we could set solver dependent surfaces / sources / numerical_config here
    # just as we have got solver dependent optical properties.
    
    for wavelength in boosted_sensor_dict.get_unique_solvers():
        medium = {
            'cloud': boosted_optical_properties[wavelength],
            'rayleigh':rayleigh_scattering[wavelength]
         }
        config = at3d.configuration.get_config()
        boosted_solvers_dict.add_solver(
            wavelength,
            at3d.solver.RTE(
                numerical_params=config,
                surface=at3d.surface.lambertian(0.0),
                source=at3d.source.solar(wavelength, 0.5,0.0),
                medium=medium,
                num_stokes=1#sensor_dict.get_minimum_stokes()[wavelength],
            )                   
       )
    
    # solve the 4 RTEs in parallel AND get the measurements.
    boosted_sensor_dict.get_measurements(boosted_solvers_dict, n_jobs=4, verbose=True)
    return boosted_sensor_dict
```

```python
def get_boosted_sensors(boost):
    #define data with reff boosted
    boosted_cloud_scatterer = cloud_scatterer.copy()
    boosted_cloud_scatterer.reff.data = boosted_cloud_scatterer.reff.data + boost
    return get_sensors_from_data(boosted_cloud_scatterer)
```

```python
#Make heat maps
boosts = [1, 5, 10]
boosted_sensor_dicts = []
for boost in boosts:
    boosted_sensor_dicts.append(get_boosted_sensors(boost))
```

```python
def plot_images(bsds, bounded=False, min=None, max=None, points=None):
    for instrument in sensor_dict:
        sensor_images = sensor_dict.get_images(instrument)
        boosted_sensor_images_array = []
        for dict in bsds:
            boosted_sensor_images_array.append(dict.get_images(instrument))
        transpose = [[row[i] for row in boosted_sensor_images_array] for i in range(len(boosted_sensor_images_array[0]))]
        for sensor, boosted_sensors in zip(sensor_images, transpose):
            fig, axs = py.subplots(1, len(boosts), figsize=(12, 4))  # Create a figure with subplots side by side
            for i in range(len(boosted_sensors)):
                # Plot sensor 1 on the first subplot
                difference = sensor.copy()
                difference.I.data = sensor.I.data-boosted_sensors[i].I.data
                if bounded:
                    difference.I.T.plot(ax=axs[i], vmin=min, vmax=max)
                else:
                    difference.I.T.plot(ax=axs[i])
                axs[i].set_title(instrument + " Difference Heat Map : Boost = " + str(boosts[i]))
            
            py.tight_layout()  # Adjust layout for better spacing
            py.show()  # Show the plot
```

```python
plot_images(boosted_sensor_dicts, True, -0.04, 0.04)
```

```python
def get_single_boosted_data(boost, points):
    #define data with single reff boosted
    boosted_cloud_scatterer = cloud_scatterer.copy()
    for point in points:
        boosted_cloud_scatterer.reff.data[point[0], point[1], point[2]] += boost
    return get_sensors_from_data(boosted_cloud_scatterer)

def get_point():
    # Get dimensions of the array
    dim1, dim2, dim3 = cloud_scatterer.reff.shape
    
    # Randomly choose indices
    idx1 = np.random.choice(dim1)
    idx2 = np.random.choice(dim2)
    idx3 = np.random.choice(dim3)
    return [idx1, idx2, idx3]
```

```python
#Make heat maps
boosts = [1, 5]
boosted_sensor_dicts = []
points = []
for i in range(100):
    points.append(get_point())
for boost in boosts:
    boosted_sensor_dicts.append(get_single_boosted_data(boost, points))
```

```python
#plot_images(boosted_sensor_dicts)
```

```python
def plot_scatter(fig, zenith, azimuth):
    # Basic scatter plot
    ax = fig.add_subplot(133, projection='3d')
    
    # Scatter plot with colormap mapping to z value
    for point in points:
        scatter = ax.scatter(point[0], point[1], point[2], c="red", s=50)
    
    # Colorbar
    
    ax.set_xlabel('X Label')
    ax.set_ylabel('Y Label')
    ax.set_zlabel('Z Label')
    ax.view_init(azimuth, zenith)
    py.title('4D Scatter Plot')
    py.show()
```

```python
def plot_MSPI_with_scatter():
    sensor_images = sensor_dict.get_images("MSPI")
    boosted_sensor_images_array = []
    for dict in boosted_sensor_dicts:
        boosted_sensor_images_array.append(dict.get_images("MSPI"))
    for i in range(len(sensor_images)):
        fig, axs = py.subplots(1, len(boosts)+1, figsize=(15, 4))  # Create a figure with subplots side by side
        sensor = sensor_images[i]
        for j in range(len(boosts)):
            difference = sensor.copy()
            difference.I.data = sensor.I.data-boosted_sensor_images_array[j][i].I.data
            difference.I.T.plot(ax=axs[j])
            axs[j].set_title("MSPI Difference Heat Map : Boost = " + str(boosts[j]))
        plot_scatter(fig, sensor_zenith_list[i], sensor_azimuth_list[i])
        py.tight_layout()  # Adjust layout for better spacing
        py.show()  # Show the plot
```

```python
plot_MSPI_with_scatter()
```
